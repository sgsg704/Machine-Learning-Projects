```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# i guess i will not be taking date into consideration so i will remove 
df.drop(["Date"],inplace=True,axis=1)
#once i use the function inplace = True that means it will be valid through entire notebook and changes will be permanent 
#axis= 1 signifies that particluar axis from where we are removing that is columns 
```

```python
#lets declare a vaiable called predictions so it basically means for how many days we are going to predict the bitcoin price 
predictionDays = 30
# Create another column shifted 'n'  units up
df['Prediction'] = df[['Price']].shift(-predictionDays)
```

```python
df
```

```python
df.tail()
```

```python
#now we will create the data set into an independent data set 
# we will convert the dataframe into numpy array 
x=np.array(df.drop(["Prediction"],axis=1))
```

```python
print(x)
```

```python
#now what we will do is that first we will remove the last n rows where n is the prediction days 
x=x[:len(df)-predictionDays]
print(x)
```

```python
y=np.array(df["Prediction"])

y=y[:-predictionDays]
print(y)
```

```python
#spliting the data inot training data and testing data 
from sklearn.model_selection import train_test_split

```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
predictionDays_array = np.array(df.drop(['Prediction'],1))[-predictionDays:]
print(predictionDays_array)
```

```python
from sklearn.svm import SVR
model=SVR(C=1000.0, cache_size=200, coef0=0.0, degree=3, epsilon=0.1, gamma=1e-05,
    kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
print(y_test)
```

```python
accuracy_score=model.score(x_test,y_test)
```

```python
accuracy_score
```

```python
# Print the model predictions for the next 30 days
svm_prediction = model.predict(predictionDays_array)
print(svm_prediction)
print()
#Print the actual price for bitcoin for last 30 days
print(df.tail(predictionDays))
```
